```python
# This exercise focuses on integrating data from two separate datasets: student.csv and marks.csv.
# Each dataset contains different attributes related to students, including demographic and academic details.
# Both datasets share a common identifier: Student_id.
# The goal is to load both datasets using pandas and merge them into a single DataFrame
# to enable unified analysis of each studentâ€™s complete profile.

import pandas as pd
dataset1 = "https://raw.githubusercontent.com/TrainingByPackt/Data-Science-with-Python/refs/heads/master/Chapter01/Data/student.csv"
dataset2 = "https://raw.githubusercontent.com/TrainingByPackt/Data-Science-with-Python/refs/heads/master/Chapter01/Data/mark.csv"
df1 = pd.read_csv(dataset1, header = 0)
df2 = pd.read_csv(dataset2, header = 0)
```

```python
df = pd.merge(df1, df2, on = 'Student_id') # Same column on both DF
df.head(10)
```
