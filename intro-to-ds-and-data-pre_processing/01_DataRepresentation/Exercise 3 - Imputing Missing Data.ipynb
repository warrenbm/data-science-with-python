```python
import pandas as pd
dataset = 'https://raw.githubusercontent.com/TrainingByPackt/Data-Science-with-Python/refs/heads/master/Chapter01/Data/Banking_Marketing.csv'
df = pd.read_csv(dataset, header=0)
```

```python
df
```

```python
mean_age = df.age.mean()
print(mean_age)
```

```python
df.age.fillna(mean_age, inplace=True)
```

```python
df.age
```

```python
median_duration = df.duration.median()
print(median_duration)
```

```python
df.duration.fillna(median_duration,inplace=True)
```

```python
df.duration
```

```python
print(df['duration'].isna().sum())
```

```python
mode_contact = df.contact.mode()[0]
print(mode_contact)
```

```python
df.contact
```

```python
df.contact.fillna(mode_contact,inplace=True)
```
