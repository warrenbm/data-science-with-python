```python
import pandas as pd
dataset = 'https://raw.githubusercontent.com/TrainingByPackt/Data-Science-with-Python/refs/heads/master/Chapter01/Data/Student_bucketing.csv'
df = pd.read_csv(dataset, header=0)
df
```

```python
df['bucket']=pd.cut(df['marks'],5,labels=['Poor','Below_average','Average','Above_Average','Excellent'])
df
```

```python
bucketed, bins = pd.cut(df['marks'],5, labels=['Poor','Below_Average','Average','Above_Average','Excellent'], retbins=True)
print("Borne des intervalles :",bins)
```

```python
df['bucket'] = bucketed
df['marks_bin_range'] = pd.cut(df['marks'], 5)
df[['marks','bucket', 'marks_bin_range']].head(10)
```
