```python
# This exercise focuses on converting categorical data into numerical form using One-Hot Encoding.
# The dataset used is Banking_Marketing.csv, which contains information about customers and their responses to marketing campaigns.
# The goal is to load the dataset using pandas and transform all categorical features (e.g., job, marital status, education)
# into a numerical format by creating binary columns for each category.
# This preprocessing step is crucial to prepare the data for machine learning models, which require numeric input.

import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
dataset = "https://raw.githubusercontent.com/TrainingByPackt/Data-Science-with-Python/refs/heads/master/Chapter01/Data/Banking_Marketing.csv"
df = pd.read_csv(dataset, header=0)
df
```

```python
df = df.dropna()
```

```python
data_column_category = df.select_dtypes(exclude=[np.number]).columns
data_column_category
```

```python
df[data_column_category].head()
```

```python
#performing label encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

for i in data_column_category:
    df[i] = label_encoder.fit_transform(df[i])
print("Label Encoded Data: ")
df.head()
```

```python
df[data_column_category].head()
```

```python
#Performing Onehot Encoding
onehot_encoder = OneHotEncoder(sparse=False)
onehot_encoded = onehot_encoder.fit_transform(df[data_column_category])
```

```python
#Creating a dataframe with encoded data with new column name
onehot_encoded_frame = pd.DataFrame(onehot_encoded, columns=onehot_encoder.get_feature_names(data_column_category))
onehot_encoded_frame.head()
```
