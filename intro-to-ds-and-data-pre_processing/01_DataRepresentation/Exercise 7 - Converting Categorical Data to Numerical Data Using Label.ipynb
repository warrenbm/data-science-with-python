```python
# This exercise focuses on converting categorical data into numerical values using label encoding.
# The dataset used is Banking_Marketing.csv, which contains customer information including
# both numerical and categorical attributes (e.g., job, marital status, education).
# The goal is to load the dataset using pandas and convert all categorical fields
# into numeric values using the LabelEncoder from sklearn.
# This transformation is crucial for enabling the use of these features in machine learning models,
# which typically require numerical inputs.

import pandas as pd
import numpy as np
dataset = "https://raw.githubusercontent.com/TrainingByPackt/Data-Science-with-Python/refs/heads/master/Chapter01/Data/Banking_Marketing.csv"
df = pd.read_csv(dataset, header =0)
```

```python
df.isna().sum()
```

```python
df = df.dropna()
df
```

```python
data_column_category = df.select_dtypes(exclude=[np.number]).columns
data_column_category
```

```python
df[data_column_category].head()
```

```python
df[data_column_category].dtypes
```

```python
#import the LaberEncoder class
from sklearn.preprocessing import LabelEncoder
#Creating the object instance
label_encoder = LabelEncoder()

for i in data_column_category:
    df[i] = label_encoder.fit_transform(df[i])
    print("Label Encoded Data: ")
    df.head()
```
